```python
import numpy as np
#from scipy import optimize
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
def line_search_secant(function, initial_point, threshold = 1e-5, max_iteration = 1000):
    point_storage = []
    point = initial_point
    
    point_storage.append(point)
    point_two_before = point
    
    point_storage.append(point + 0.5)
    point_one_before = point + 0.5
    
    point_storage.append(point + 1)
    point = point + 1
    
    i = 1
    while True:
        i+=1
        point = point_one_before - function(point_one_before)*(point_one_before-point_two_before)/(function(point_one_before)-function(point_two_before))
        
        point_storage.append(point)
        compare_value = np.linalg.norm(point - point_one_before)
        
        if ((compare_value < threshold) and (point>=0)) or (i >max_iteration):
            #print(compare_value)
            #print("Convergence at {} iterations".format(i))
            return point
            #break
        
        point_two_before = point_one_before
        point_one_before = point
    
```

```python
def obj_func(x):
    x1, x2 = x
    return 1 + 2*x1 + 3*(x1**2+x2**2)+4*x1*x2

def gradient_obj_func(x):
    x1, x2 = x
    f_x1 = 6*x1 + 4*x2 + 2
    f_x2 = 4*x1 + 6*x2
    return np.array([f_x1, f_x2 ])

def gradient_alpha(alpha):
    value = -12*point[0] -8*point[1] + 3*(-8*point[0] - 12*point[1])\
    *(-alpha*(4*point[0] + 6*point[1])+point[1]) + (-4*point[0] - 6*point[1])\
    *(-4*alpha*(6*point[0] + 4*point[1] + 2) + 4*point[0])\
    + (-alpha*(4*point[0] + 6*point[1]) + point[1])*(-24*point[0]-16*point[1] -8)\
    + 3*(-alpha*(6*point[0]+4*point[1] + 2)+point[0])*(-12*point[0]-8*point[1]-4)-4
    return value
```

```python
# Hyperparameter
threshold = 0.0003     # convergence threshold

# steepest Descent
point = np.array([-2.0, 2.0])   
x, y = point
point_storage = []
point_storage.append(point)


i = 0
while True:
    i += 1
    print(point)
    gradient = gradient_obj_func(point)

    step_size = line_search_secant(gradient_alpha, 0)
    point = point - step_size * gradient_obj_func(point)
    point_storage.append(point)

    print("Point History : {}".format(point_storage))
    print("Next Point : {}".format(point))
    print('Present step_size : {}\n'.format(step_size))
    compare_value = np.linalg.norm(point_storage[i] - point_storage[i-1])

    if compare_value < threshold:
        print("Convergence Value : {}".format(compare_value))
        print("Convergence at {} iterations".format(i))
        break
        
```
